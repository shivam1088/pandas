```python
import numpy as np

data = [23, 34, 23, 2, 45, 67, 34, 6, 3, 80, 3, 48, 3, 4, -67,500]

def detect_outliers(data):
    outliers = []  # Initialize outliers list inside the function
    threshold = 3  # Z-score threshold for identifying outliers
    mean = np.mean(data)  # Calculate the mean of the data
    std = np.std(data)  # Calculate the standard deviation of the data

    # Loop through the data and calculate Z-scores
    for i in data:
        z_score = (i - mean) / std
        if np.abs(z_score) > threshold:
            outliers.append(i)  # Append the outlier to the list

    return outliers

# Detect outliers
res = detect_outliers(data)
print(res)
```
