```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
x=['c','c++','java','python','git']
y=[70,20,30,40,50]
plt.xlabel('Languge',fontsize=40)
plt.ylabel('No.',fontsize=50)
plt.title('Ranking system',fontsize=40)
c=['r','g','b','black','y']
plt.bar(x,y,color=c,width=0.4,align='edge',edgecolor='b',linewidth=5,linestyle=':',label='Popularity')
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5,67,80]
y=[10,20,30,40,50,60,70]
s=[120,320,140,50,160,120,145]
plt.scatter(x,y,color='b',edgecolor='r',marker='*',sizes=s,cmap='Purples',linewidth=2)
plt.colorbar()
plt.show()
```

```python
# Histogram plot
```

```python
import numpy as np
import random
x=np.random.randint(10,60,(100))
plt.title('histogram plot',fontsize=50)
plt.xlabel('rank1')
plt.ylabel('rank2')
plt.hist(x,color='r',edgecolor='black',align='right',histtype='stepfilled',log=True,rwidth=3,label='python_histogram_plot')
plt.legend()
plt.show()
```

```python
#PIE PLOT
```

```python
x=['c','c++','java','python','git']
y=[70,20,130,110,50]
ex=[0,0.1,0.0,0,0.1]
c=['r','g','b','m','pink','y']
plt.title('wscubetech')
plt.pie(y,explode=ex,labels=x,colors=c,shadow=False,autopct='% 0.1f %%',radius=0.8,labeldistance=1.3,textprops={'fontsize':10},wedgeprops={'linewidth':5,'width':0.5,'edgecolor':'black'},center=(4,8),rotatelabels=False)
plt.legend(loc=1)
plt.show()
```
